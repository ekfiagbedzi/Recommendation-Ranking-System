```python
def reqs(api_path: str=""):
    response = requests.get("https://api.stackexchange.com{}".format(api_path))
    return response.json()
```

```python
curious = reqs("/2.3/questions?page=25&pagesize=100&order=desc&sort=activity&site=stackoverflow")
```
